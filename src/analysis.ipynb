# Result analysis

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

labels = ['PyTorch', "FoldConstant", "EliminateCommonSubexpr", "CustomizedEliminateCommonSubexpr", "RemoveUnusedFunctions", "ToBasicBlockNormalForm", 
          "CombineParallelBatchMatmul", "Sequential", "AutoTVM", "ANSOR","ANSOR on CPBM"]
results = [67.52, 66.64, 67.74, 66.44, 66.84, 67.37, 66.24, 67.86, 73.59, 61.66, 61.57]
gain = [(67.52 - time)/67.52 * 100 for time in results]
df = pd.DataFrame({'opt':labels, 'time':gain})
ax = df.plot.bar(x='opt', y='time', rot=0)

plt.title("BERT optimization latency time")
plt.xlabel("Optimization")
plt.ylabel("Time gain compared with PyTorch (%)")
plt.xticks(rotation='vertical')
```
